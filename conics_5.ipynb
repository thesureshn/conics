```python
#given Eccentricity is e = 3/5 and foci is 6 centre is origin an ellipse
import numpy as np
import matplotlib.pyplot as plt
from coeffs import *
```

```python
# e = c/a = 3/5 

O = np.array([0.,0.])
u=O[0]     #x-position of the center
v=O[1]    #y-position of the center
b=4.     #radius on the x-axis
a=5.    #radius on the y-axis
```

```python
A = np.array([5,0]) # point on major axis
B = np.array([0,4]) #point on Minor axis
C = np.array([-5,0]) # point on major axis
D = np.array([0,-4]) #point on Minor axis

X_AB = line_gen(A,B)
X_BC = line_gen(B,C)
X_CD = line_gen(C,D)
X_DA = line_gen(D,A)
```

```python
t = np.linspace(0, 2*np.pi, 100)
Ell = np.array([u+a*np.cos(t) , v+b*np.sin(t)])
#plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.plot(Ell[0,:],Ell[1,:],label='$ellipse$')
plt.grid(color='lightgray',linestyle='--')
plt.plot(X_AB[0,:],X_AB[1,:],label='$AB$')
plt.plot(X_BC[0,:],X_BC[1,:],label='$BC$')
plt.plot(X_CD[0,:],X_CD[1,:],label='$CD$')
plt.plot(X_DA[0,:],X_DA[1,:],label='$DA$')
plt.plot(O[0],O[1], 'o',label='$Centre$')
plt.legend(loc="best")
plt.show()
```
