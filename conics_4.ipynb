```python
import numpy as np
import matplotlib.pyplot as plt
from coeffs import *
```

```python
# Generating Co-ordinates for Parabola
len = 100
y =np.linspace(-5,5,len)
x = (1/3)*(y**2)
x_par1= np.vstack((x,y)) # Contains Co-Ordinates of Parabola y^2=3x

O = np.array([0,-6])
R = np.sqrt(37)
theta = np.linspace(0,2*np.pi,len)
x_circ = np.zeros((2,len))
x_circ[0,:] = R*np.cos(theta)
x_circ[1,:] = R*np.sin(theta)
x_circ = (x_circ.T + O).T
```

```python
m = np.array([2,2])
A = O + 5*(m/np.linalg.norm(m))

x_OA = line_gen(O,A)
```

```python
plt.plot(x_par1[0,:],x_par1[1,:], label='$Parabola$')
plt.plot(x_circ[0,:],x_circ[1,:],label='$incircle$')
plt.plot(x_OA[0,:],x_OA[1,:],label='$OP$')

plt.plot(O[0], O[1], 'o')
plt.text(O[0] * (1 + 0.1), O[1] * (1 - 0.1) , 'O')
plt.plot(A[0], A[1], 'o')
plt.text(A[0] * (1 + 0.1), A[1] * (1 - 0.1) , 'P')



plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.axis('equal')
plt.legend()
#plt.axes().set_aspect('equal')
plt.show()
```
